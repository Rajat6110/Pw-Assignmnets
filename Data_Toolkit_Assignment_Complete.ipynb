### Question 7
Clean 'Phone' column: remove non-numeric characters and convert to numeric; show dtypes.

```python
# Detect phone-like column (case-insensitive)
phone_col = None
for c in df.columns:
    if c.strip().lower().startswith('phone'):
        phone_col = c
        break
print('Detected phone column:', phone_col)
if phone_col:
    df['Phone_clean'] = df[phone_col].astype(str).str.replace(r'[^0-9]', '', regex=True)
    df['Phone_clean'] = pd.to_numeric(df['Phone_clean'], errors='coerce')
print('\nDtypes:') 
print(df.dtypes)
if phone_col in df.columns:
    display(df[[phone_col,'Phone_clean']].head(5))
else:
    print('No phone column found.')
```

### Question 8
Read data.csv skipping first 50 rows; only read Last Name, Gender, Email, Phone, Salary; show first 10 rows; show last 5 Salary values.

```python
try:
    df_skip = pd.read_csv(r"/mnt/data/People Data.csv", skiprows=50, low_memory=False)
    cand = {}
    for want in ['Last Name','Gender','Email','Phone','Salary']:
        for c in df_skip.columns:
            if c.strip().lower().replace(' ','') == want.strip().lower().replace(' ',''):
                cand[want]=c; break
    print('Found columns:', cand)
    sel = list(cand.values())
    if sel:
        display(df_skip[sel].head(10))
    if 'Salary' in cand:
        print('\nSalary last 5:') 
        print(df_skip[cand['Salary']].tail(5))
except Exception as e:
    print('Error reading skiprows=50:', e)
```

### Question 9
Filter rows where Last Name contains 'Duke', Gender contains 'Female', Salary < 85000.

```python
lname = next((c for c in df.columns if 'last' in c.lower() and 'name' in c.lower()), None)
gender = next((c for c in df.columns if 'gender' in c.lower()), None)
salary = next((c for c in df.columns if 'salary' in c.lower()), None)
print(lname, gender, salary)
if all([lname,gender,salary]):
    res = df[df[lname].astype(str).str.contains('duke',case=False,na=False) & df[gender].astype(str).str.contains('female',case=False,na=False) & (pd.to_numeric(df[salary],errors='coerce')<85000)]
    print('Matches:', res.shape[0])
    display(res.head(10))
else:
    print('Required columns not found.')
```

### Question 12
Drop Email, Phone, Date of birth columns, then drop rows with missing values, print final shape and head.

```python
drops = []
for c in df.columns:
    if any(k in c.lower() for k in ['email','phone']) or ('date' in c.lower() and 'birth' in c.lower()) or c.strip().lower()=='dob':
        drops.append(c)
drops = list(dict.fromkeys(drops))
print('Dropping:', drops)
df2 = df.drop(columns=[c for c in drops if c in df.columns], errors='ignore').dropna()
print('Final shape:', df2.shape)
df2.head()
```

### Question 17
Bokeh sine wave saved as HTML file.

```python
try:
    from bokeh.plotting import figure, output_file, save
    import numpy as np
    x = np.linspace(0,2*np.pi,200); y=np.sin(x)
    p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)', width=700, height=300)
    p.line(x,y)
    output_file('/mnt/data/bokeh_sine_wave.html')
    save(p)
    print('Saved /mnt/data/bokeh_sine_wave.html')
except Exception as e:
    print('Bokeh not available or failed:', e)
```

### Question 18
Bokeh random categorical bar chart saved as HTML file.

```python
try:
    from bokeh.plotting import figure, output_file, save
    from bokeh.models import ColumnDataSource, HoverTool
    cats=['A','B','C','D','E']; vals = np.random.randint(1,100,len(cats))
    source = ColumnDataSource({'cats':cats,'vals':vals})
    p = figure(x_range=cats, title='Random Categorical Bar Chart', width=700, height=350)
    p.vbar(x='cats', top='vals', width=0.6, source=source)
    p.add_tools(HoverTool(tooltips=[('cat','@cats'),('val','@vals')]))
    output_file('/mnt/data/bokeh_bar_chart.html'); save(p)
    print('Saved /mnt/data/bokeh_bar_chart.html')
except Exception as e:
    print('Bokeh failed:', e)
```

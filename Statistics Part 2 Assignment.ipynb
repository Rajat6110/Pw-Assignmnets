## Theoretical Questions

### 1. What is hypothesis testing in statistics?

> Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on a sample of data. It involves setting up a null hypothesis (a statement of no effect or no difference) and an alternative hypothesis, and then using sample data to determine whether there is enough evidence to reject the null hypothesis.

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

> The **null hypothesis ($H_0$)** represents a statement of no change, no difference, or no relationship in the population. It's the default assumption that we seek to challenge. The **alternative hypothesis ($H_a$ or $H_1$)** is the opposite of the null hypothesis; it represents the presence of an effect, a difference, or a relationship. The goal of hypothesis testing is to see if there's enough evidence in the sample data to reject the null hypothesis in favor of the alternative.

### 3. What is the significance level in hypothesis testing, and why is it important?

> The **significance level**, denoted by **alpha (α)**, is the probability of rejecting the null hypothesis when it is actually true. It's the threshold for deciding whether a result is statistically significant. A common significance level is 0.05 (or 5%), meaning there's a 5% risk of concluding that a difference exists when there is no actual difference. It's important because it quantifies the risk of a Type I error (a false positive) that the researcher is willing to accept.

### 4. What does a P-value represent in hypothesis testing?

> The **P-value** (probability value) is the probability of obtaining test results at least as extreme as the results actually observed, assuming the null hypothesis is correct. A small P-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, so you reject the null hypothesis. A large P-value (> 0.05) indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis.

### 5. How do you interpret the P-value in hypothesis testing?

> To interpret the P-value, you compare it to the predetermined significance level (α):
> * **If P-value ≤ α:** You reject the null hypothesis. This means the observed data is statistically significant, and there is enough evidence to support the alternative hypothesis.
> * **If P-value > α:** You fail to reject the null hypothesis. This means the observed data is not statistically significant, and there isn't enough evidence to support the alternative hypothesis.

### 6. What are Type 1 and Type 2 errors in hypothesis testing?

> * **Type I Error (False Positive):** This occurs when you **reject a true null hypothesis**. The probability of a Type I error is equal to the significance level (α).
> * **Type II Error (False Negative):** This occurs when you **fail to reject a false null hypothesis**. The probability of a Type II error is denoted by beta (β).

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

> * A **one-tailed test** is used when the alternative hypothesis specifies a direction (e.g., the new drug is *better* than the old one). The critical region for rejection is located in only one tail of the distribution (either left or right).
> * A **two-tailed test** is used when the alternative hypothesis does not specify a direction (e.g., there is *a difference* between the new and old drug, but it could be better or worse). The critical region is split between both tails of the distribution.

### 8. What is the Z-test, and when is it used in hypothesis testing?

> The **Z-test** is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large (typically n > 30). It is based on the standard normal distribution (Z-distribution).

### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

> The Z-score (or Z-statistic) is calculated using the formula: $$ Z = \frac{(\bar{x} - \mu)}{(\frac{\sigma}{\sqrt{n}})} $$
> where:
> * $\bar{x}$ is the sample mean
> * $\mu$ is the population mean
> * $\sigma$ is the population standard deviation
> * $n$ is the sample size
>
> The Z-score represents how many standard deviations a data point (or sample mean) is from the population mean.

### 10. What is the T-distribution, and when should it be used instead of the normal distribution?

> The **T-distribution** (or Student's t-distribution) is a probability distribution that is used to estimate population parameters when the sample size is small and/or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, meaning it is more prone to producing values that fall far from its mean. It should be used instead of the normal distribution in these scenarios.

### 11. What is the difference between a Z-test and a T-test?

> The main difference lies in the conditions under which they are used:
> * **Z-test:** Used when the population standard deviation ($\sigma$) is **known** and the sample size is large (n > 30).
> * **T-test:** Used when the population standard deviation ($\sigma$) is **unknown** and is estimated from the sample standard deviation (s), or when the sample size is small (n < 30).

### 12. What is the T-test, and how is it used in hypothesis testing?

> A **T-test** is a type of inferential statistic used to determine if there is a significant difference between the means of two groups, which may be related in certain features. It is used in hypothesis testing to compare a sample mean to a known population mean (one-sample t-test), to compare the means of two independent groups (independent t-test), or to compare the means of the same group at different times (paired t-test).

### 13. What is the relationship between Z-test and T-test in hypothesis testing?

> The T-distribution is closely related to the normal distribution. As the sample size (and thus the degrees of freedom) increases, the T-distribution approaches the standard normal distribution (Z-distribution). For large sample sizes (e.g., > 30), the T-test and Z-test will produce very similar results. The T-test is more versatile as it can be used for small samples, while the Z-test requires a large sample or known population variance.

### 14. What is a confidence interval, and how is it used to interpret statistical results?

> A **confidence interval** is a range of values, derived from sample statistics, that is likely to contain the value of an unknown population parameter. For example, a 95% confidence interval for a population mean implies that if you were to take 100 different samples and compute a 95% confidence interval for each sample, then approximately 95 of the 100 confidence intervals will contain the true population mean. It provides a measure of the uncertainty associated with a sample estimate.

### 15. What is the margin of error, and how does it affect the confidence interval?

> The **margin of error** is a statistic expressing the amount of random sampling error in the results of a survey. It is half the width of the confidence interval. A larger margin of error means a wider confidence interval, indicating less precision in the estimate of the population parameter. The margin of error is affected by the sample size, the variability of the population, and the desired confidence level.

### 16. How is Bayes' Theorem used in statistics, and what is its significance?

> **Bayes' Theorem** describes the probability of an event, based on prior knowledge of conditions that might be related to the event. The formula is: $$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$
> In statistics, it is the foundation of Bayesian inference, where it's used to update the probability for a hypothesis as more evidence or information becomes available. It's significant because it provides a mathematical framework for combining prior beliefs with observed data to arrive at a more informed posterior probability.

### 17. What is the Chi-square distribution, and when is it used?

> The **Chi-square ($\chi^2$) distribution** is a continuous probability distribution that is widely used in hypothesis testing. It is the distribution of a sum of the squares of k independent standard normal random variables. It is used in:
> * **Goodness-of-fit tests:** To determine if sample data matches a population.
> * **Tests of independence:** To test for the independence of two categorical variables.
> * **Tests for variance:** To test a hypothesis about a single population variance.

### 18. What is the Chi-square goodness of fit test, and how is it applied?

> The **Chi-square goodness of fit test** is a statistical hypothesis test used to determine whether a variable is likely to come from a specified distribution or not. It is applied by comparing the observed frequencies in different categories of a categorical variable to the expected frequencies that would be present if the null hypothesis were true. A significant result suggests that the observed data does not fit the expected distribution.

### 19. What is the F-distribution, and when is it used in hypothesis testing?

> The **F-distribution** is a continuous probability distribution that arises frequently as the null distribution of a test statistic, most notably in the Analysis of Variance (ANOVA). It is the ratio of two independent chi-square variables divided by their respective degrees of freedom. It is used in hypothesis testing to compare the variances of two populations (F-test) or to compare the means of three or more groups (ANOVA).

### 20. What is an ANOVA test, and what are its assumptions?

> **Analysis of Variance (ANOVA)** is a statistical test used to analyze the differences among group means in a sample. It assesses whether the means of several groups are equal. Key assumptions for ANOVA are:
> * **Independence:** The observations are independent of one another.
> * **Normality:** The data in each group are normally distributed.
> * **Homogeneity of variances (homoscedasticity):** The variances of the groups are equal.

### 21. What are the different types of ANOVA tests?

> * **One-way ANOVA:** Used to test for differences among the means of two or more independent groups based on one independent variable (factor).
> * **Two-way ANOVA:** Used to evaluate the effect of two different independent variables on one continuous dependent variable. It also assesses the interaction between the two factors.
> * **N-way ANOVA:** An extension of the two-way ANOVA, for studying the effects of three or more independent variables.
> * **Repeated Measures ANOVA:** Used when the same subjects are used for each treatment (e.g., measuring the same person at different points in time).

### 22. What is the F-test, and how does it relate to hypothesis testing?

> The **F-test** is a statistical test that uses the F-distribution. In hypothesis testing, it is most commonly used to compare the variances of two populations or to compare the means of multiple groups (as part of an ANOVA). The F-statistic is a ratio of two variances. If the ratio is close to 1, it suggests that the two variances are equal, supporting the null hypothesis.

### 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#%pip install statsmodels

import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data
sample_data = [2.5, 3.2, 2.8, 3.5, 3.0, 2.9, 3.1, 2.7, 3.3, 2.8]
population_mean = 3.0

# Perform the Z-test
z_statistic, p_value = ztest(sample_data, value=population_mean)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

### 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np

def one_sample_z_test(sample_data, pop_mean, pop_std):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    # P-value can be calculated using scipy.stats.norm.sf
    from scipy.stats import norm
    p_value = 2 * norm.sf(abs(z_score)) # Two-tailed test
    return z_score, p_value

# Example usage
sample = [10.2, 9.8, 10.5, 10.1, 9.9, 10.3, 9.7]
population_mean = 10.0
population_std = 0.5

z, p = one_sample_z_test(sample, population_mean, population_std)
print(f"Z-score: {z:.4f}")
print(f"P-value: {p:.4f}")
```

### 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
alpha = 0.05

# Generate Z distribution data
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Calculate critical values
z_critical = norm.ppf(1 - alpha / 2)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.fill_between(x, y, where=(x <= z_critical) & (x >= -z_critical), color='green', alpha=0.5, label='Acceptance Region')
plt.axvline(z_critical, color='k', linestyle='--', label=f'Critical Value: {z_critical:.2f}')
plt.axvline(-z_critical, color='k', linestyle='--')
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

### 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0, mu1, sigma, n, alpha):
    # H0 distribution
    x = np.linspace(mu0 - 4*sigma/np.sqrt(n), mu1 + 4*sigma/np.sqrt(n), 1000)
    h0_dist = norm(mu0, sigma/np.sqrt(n))
    
    # H1 distribution
    h1_dist = norm(mu1, sigma/np.sqrt(n))
    
    # Critical value
    critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma/np.sqrt(n))
    
    # Type 1 Error (alpha)
    type1_error_prob = 1 - h0_dist.cdf(critical_value)
    
    # Type 2 Error (beta)
    type2_error_prob = h1_dist.cdf(critical_value)
    
    plt.figure(figsize=(12, 7))
    plt.plot(x, h0_dist.pdf(x), label='H0: Null Hypothesis')
    plt.plot(x, h1_dist.pdf(x), label='H1: Alternative Hypothesis')

    # Shade Type 1 Error
    x_fill1 = np.linspace(critical_value, mu0 + 4*sigma/np.sqrt(n), 100)
    plt.fill_between(x_fill1, h0_dist.pdf(x_fill1), color='red', alpha=0.5, label=f'Type 1 Error (α) = {type1_error_prob:.2f}')
    
    # Shade Type 2 Error
    x_fill2 = np.linspace(mu0 - 4*sigma/np.sqrt(n), critical_value, 100)
    plt.fill_between(x_fill2, h1_dist.pdf(x_fill2), color='blue', alpha=0.5, label=f'Type 2 Error (β) = {type2_error_prob:.2f}')
    
    plt.axvline(critical_value, color='k', linestyle='--', label='Critical Value')
    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return type1_error_prob, type2_error_prob

# Example usage
visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

### 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval_mean(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data) # Standard error of the mean
    h = std_err * stats.t.ppf((1 + confidence) / 2., n-1)
    return mean, mean - h, mean + h

# Example
sample = [12, 12, 13, 13, 15, 16, 17, 18, 22, 25]
mean, lower_bound, upper_bound = confidence_interval_mean(sample)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
print("Significance: We are 95% confident that the true population mean lies between",
f"{lower_bound:.2f} and {upper_bound:.2f}.")
```

### 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence_level=0.95):
    n = len(data)
    std_dev = np.std(data, ddof=1) # Sample standard deviation
    # Get the critical value (z-score for large samples, t-score for small)
    if n >= 30:
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    
    margin_of_error = critical_value * (std_dev / np.sqrt(n))
    return margin_of_error

# Sample data
sample_data = [65, 68, 70, 71, 73, 75, 76, 78, 80, 82, 85, 88, 90, 92, 95]
moe = calculate_margin_of_error(sample_data)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Margin of Error (95% Confidence): {moe:.2f}")
```

### 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B_given_not_A):
    # P(not A)
    prior_not_A = 1 - prior_A
    
    # P(B)
    prob_B = (prob_B_given_A * prior_A) + (prob_B_given_not_A * prior_not_A)
    
    # P(A|B) using Bayes' Theorem
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B
    
    return posterior_A_given_B

# Example: Medical test
# A: Person has the disease
# B: Test result is positive

# P(A): Prior probability of having the disease (prevalence)
prior_disease = 0.01  # 1% of the population has the disease

# P(B|A): Probability of positive test given disease (sensitivity)
prob_positive_given_disease = 0.95 # Test is 95% sensitive

# P(B|not A): Probability of positive test given no disease (false positive rate)
prob_positive_given_no_disease = 0.10 # Test has a 10% false positive rate

# Calculate the posterior probability: P(Disease|Positive Test)
posterior_disease_given_positive = bayes_theorem(prior_disease,
                                                 prob_positive_given_disease,
                                                 prob_positive_given_no_disease)

print(f"Prior probability of disease: {prior_disease:.4f}")
print(f"Probability of positive test given disease: {prob_positive_given_disease:.4f}")
print(f"Probability of positive test given no disease: {prob_positive_given_no_disease:.4f}")
print(f"Posterior probability of having the disease given a positive test: {posterior_disease_given_positive:.4f}")

print("\nProcess Explanation:")
print("1. We start with a prior belief (prior_disease) about the probability of an event (having the disease).")
print("2. We observe new evidence (a positive test result).")
print("3. We use the likelihoods (sensitivity and false positive rate) to see how probable the evidence is under the hypothesis (disease) and its alternative (no disease).")
print("4. Bayes' Theorem updates our prior belief to a posterior probability (posterior_disease_given_positive) by combining the prior with the likelihood of the evidence.")
print("In this example, despite a 95% sensitive test, if the disease is rare, a positive result doesn't guarantee a high probability of having the disease due to the relatively high false positive rate compared to the prior prevalence.")
```

### 12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed data (contingency table)
# Rows: Gender (Male, Female)
# Columns: Opinion (Agree, Disagree, Neutral)
observed_data = np.array([
    [50, 30, 20],  # Male counts
    [40, 60, 50]   # Female counts
])

# Perform Chi-square test for independence
chi2_stat, p_value, dof, expected_freq = chi2_contingency(observed_data)

print(f"Observed Frequencies:\n{observed_data}")
print(f"\nExpected Frequencies:\n{expected_freq.round(2)}")
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association (dependence) between gender and opinion.")
else:
    print("\nFail to reject the null hypothesis: There is no significant association (independence) between gender and opinion.")
```

### 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    row_totals = np.sum(observed_data, axis=1)
    col_totals = np.sum(observed_data, axis=0)
    grand_total = np.sum(observed_data)
    
    expected_frequencies = np.outer(row_totals, col_totals) / grand_total
    return expected_frequencies

# Observed data (same as previous example)
observed = np.array([
    [50, 30, 20],
    [40, 60, 50]
])

expected = calculate_expected_frequencies(observed)

print(f"Observed Frequencies:\n{observed}")
print(f"\nCalculated Expected Frequencies:\n{expected.round(2)}")

# You can also get this directly from chi2_contingency if you run the test
from scipy.stats import chi2_contingency
chi2_stat, p_value, dof, expected_from_scipy = chi2_contingency(observed)
print(f"\nExpected Frequencies (from scipy.stats.chi2_contingency):\n{expected_from_scipy.round(2)}")
```

### 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
from scipy.stats import chisquare
import numpy as np

# Observed frequencies (e.g., counts of colors in a bag of candies)
observed_frequencies = np.array([25, 18, 32, 20, 15]) # Red, Blue, Green, Yellow, Orange

# Expected frequencies (e.g., based on manufacturer's claim or a hypothesized distribution)
# Assuming a total of 110 candies and equal distribution for simplicity (22 per color)
total_observed = np.sum(observed_frequencies)
num_categories = len(observed_frequencies)
expected_frequencies = np.full(num_categories, total_observed / num_categories)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies (assuming equal distribution): {expected_frequencies.round(2)}")

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed distribution does not significantly differ from the expected distribution.")
```

### 16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

def f_test_variances(sample1, sample2, alpha=0.05):
    n1 = len(sample1)
    n2 = len(sample2)
    s1_sq = np.var(sample1, ddof=1) # Sample variance, N-1 for unbiased estimate
    s2_sq = np.var(sample2, ddof=1)
    
    # Ensure F-statistic is > 1 by putting the larger variance in the numerator
    if s1_sq >= s2_sq:
        F_stat = s1_sq / s2_sq
        df1 = n1 - 1
        df2 = n2 - 1
    else:
        F_stat = s2_sq / s1_sq
        df1 = n2 - 1
        df2 = n1 - 1
        
    # Calculate p-value (for a two-tailed test, since we put larger variance in numerator, we double the upper tail probability)
    p_value = 2 * f.sf(F_stat, df1, df2)
    
    return F_stat, p_value, df1, df2

# Sample data
np.random.seed(42)
sample_a = np.random.normal(loc=10, scale=2, size=30)
sample_b = np.random.normal(loc=12, scale=3, size=25)

F_stat, p_value, df1, df2 = f_test_variances(sample_a, sample_b)

print(f"Sample A Variance: {np.var(sample_a, ddof=1):.4f}")
print(f"Sample B Variance: {np.var(sample_b, ddof=1):.4f}")
print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom (numerator, denominator): ({df1}, {df2})")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the variances of the two samples.")
```

### 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

def check_anova_assumptions(*groups, alpha=0.05):
    print("--- Checking ANOVA Assumptions ---")

    # 1. Independence (Assumed by design, but good to state)
    print("1. Independence: Assumed by the experimental design (groups should be independent). Please ensure your data collection method satisfies this.")

    # 2. Normality (Shapiro-Wilk Test for each group)
    print("\n2. Normality Check (Shapiro-Wilk Test for each group):")
    normal_assumption_met = True
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"   - Group {i+1} (n={len(group)}): W={stat:.4f}, p={p:.4f}")
        if p < alpha:
            print(f"     -> Group {i+1} might not be normally distributed (p < {alpha}).")
            normal_assumption_met = False
        else:
            print(f"     -> Group {i+1} appears to be normally distributed (p >= {alpha}).")
    if normal_assumption_met:
        print("   Overall: Normality assumption appears to be met for all groups.")
    else:
        print("   Overall: Normality assumption might be violated for one or more groups. Consider non-parametric tests or data transformation.")

    # 3. Homogeneity of Variances (Levene's Test)
    print("\n3. Homogeneity of Variances Check (Levene's Test):")
    stat, p = stats.levene(*groups)
    print(f"   - Levene's Test: F={stat:.4f}, p={p:.4f}")
    if p < alpha:
        print(f"     -> Homogeneity of variances assumption might be violated (p < {alpha}). Consider Welch's ANOVA or data transformation.")
    else:
        print(f"     -> Homogeneity of variances assumption appears to be met (p >= {alpha}).")
    
    print("\n--- Assumption Check Complete ---")
    print("Note: These tests provide statistical evidence, but visual inspection (e.g., Q-Q plots, box plots) is also recommended.")


# Example Usage:
np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=2.1, size=30)
group3 = np.random.normal(loc=11, scale=2.5, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for a two-way ANOVA
np.random.seed(42)
data = {
    'Score': np.concatenate([
        np.random.normal(50, 5, 20), # Factor A: Level 1, Factor B: Level 1
        np.random.normal(55, 5, 20), # Factor A: Level 1, Factor B: Level 2
        np.random.normal(60, 5, 20), # Factor A: Level 2, Factor B: Level 1
        np.random.normal(65, 5, 20)  # Factor A: Level 2, Factor B: Level 2
    ]),
    'FactorA': ['A1']*40 + ['A2']*40, # Two levels for Factor A
    'FactorB': ['B1']*20 + ['B2']*20 + ['B1']*20 + ['B2']*20 # Two levels for Factor B
}
df = pd.DataFrame(data)

# Perform Two-Way ANOVA
# The formula 'Score ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)' means:
# Dependent variable ~ Main effect of FactorA + Main effect of FactorB + Interaction effect
model = ols('Score ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2) # Type 2 ANOVA for unbalanced designs if any

print("Two-Way ANOVA Results:")
print(anova_table)

# Visualize Interaction (Interaction Plot)
plt.figure(figsize=(10, 7))
sns.pointplot(x='FactorA', y='Score', hue='FactorB', data=df, dodge=True, ci=95)
plt.title('Interaction Plot of FactorA and FactorB on Score')
plt.xlabel('Factor A')
plt.ylabel('Mean Score')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\nInterpretation:")
print("Examine the P-values in the ANOVA table:")
print(" - `C(FactorA)` row for the main effect of Factor A.")
print(" - `C(FactorB)` row for the main effect of Factor B.")
print(" - `C(FactorA):C(FactorB)` row for the interaction effect. A significant p-value here indicates that the effect of one factor depends on the level of the other factor.")
print("The interaction plot visually represents how the mean score changes across levels of one factor, conditional on the levels of the other factor. Non-parallel lines suggest an interaction effect.")
```

### 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2

def chi_square_variance_test(sample_data, hypothesized_variance, alpha=0.05):
    n = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1) # Use ddof=1 for sample variance
    
    # Chi-square test statistic for variance
    chi2_stat = ((n - 1) * sample_variance) / hypothesized_variance
    
    # Degrees of freedom
    df = n - 1
    
    # Calculate p-value for a two-tailed test
    # The p-value is the probability of observing a chi-square statistic as extreme as, or more extreme than, the one calculated.
    # For a two-tailed test, we sum the probabilities in both tails.
    p_value = 2 * min(chi2.cdf(chi2_stat, df), chi2.sf(chi2_stat, df)) # cdf for lower tail, sf for upper tail
    
    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Hypothesized Population Variance: {hypothesized_variance:.4f}")
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"\nReject the null hypothesis (alpha={alpha}): The population variance is significantly different from {hypothesized_variance}.")
    else:
        print(f"\nFail to reject the null hypothesis (alpha={alpha}): There is no significant evidence that the population variance is different from {hypothesized_variance}.")

# Example Usage
np.random.seed(42)
sample_scores = np.random.normal(loc=70, scale=10, size=25) # Sample from a population with std dev 10 (variance 100)
hypothesized_pop_variance = 80 # We hypothesize the population variance is 80

chi_square_variance_test(sample_scores, hypothesized_pop_variance)
```

### 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

### 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f, norm

def f_test_and_visualize(data1, data2, alpha=0.05):
    n1 = len(data1)
    n2 = len(data2)
    s1_sq = np.var(data1, ddof=1)
    s2_sq = np.var(data2, ddof=1)

    # Ensure F_stat >= 1 for a one-tailed F-distribution lookup for p-value (common practice)
    if s1_sq >= s2_sq:
        F_stat = s1_sq / s2_sq
        df1 = n1 - 1
        df2 = n2 - 1
        sample_names = ('Dataset 1', 'Dataset 2')
    else:
        F_stat = s2_sq / s1_sq
        df1 = n2 - 1
        df2 = n1 - 1
        sample_names = ('Dataset 2', 'Dataset 1')
    
    # Calculate p-value (two-tailed test)
    p_value = 2 * f.sf(F_stat, df1, df2)
    
    print(f"Variance of {sample_names[0]}: {s1_sq:.4f}")
    print(f"Variance of {sample_names[1]}: {s2_sq:.4f}")
    print(f"F-statistic: {F_stat:.4f}")
    print(f"Degrees of Freedom: ({df1}, {df2})")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(f"\nReject the null hypothesis (alpha={alpha}): The variances of the two datasets are significantly different.")
    else:
        print(f"\nFail to reject the null hypothesis (alpha={alpha}): There is no significant difference between the variances of the two datasets.")
    
    # Visualization
    plt.figure(figsize=(12, 6))
    
    # Subplot 1: Box plots of the datasets
    plt.subplot(1, 2, 1)
    sns.boxplot(data=[data1, data2], palette='viridis')
    plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
    plt.title('Distribution of Data in Two Datasets')
    plt.ylabel('Value')
    
    # Subplot 2: F-distribution and F-statistic
    plt.subplot(1, 2, 2)
    x = np.linspace(0, f.ppf(0.999, df1, df2), 500) # Extend x-axis to cover more of the tail
    plt.plot(x, f.pdf(x, df1, df2), label=f'F-distribution (df1={df1}, df2={df2})')
    plt.axvline(F_stat, color='red', linestyle='--', label=f'Calculated F-stat: {F_stat:.2f}')
    
    # Critical value for visualization (upper tail for the F_stat >= 1 approach)
    f_critical_upper = f.ppf(1 - alpha/2, df1, df2) # For two-tailed test, divide alpha by 2
    plt.axvline(f_critical_upper, color='green', linestyle=':', label=f'Critical F-value (α/2={alpha/2:.3f}): {f_critical_upper:.2f}')
    
    # Shade rejection region
    x_reject = np.linspace(f_critical_upper, x.max(), 100)
    plt.fill_between(x_reject, f.pdf(x_reject, df1, df2), color='orange', alpha=0.3, label='Rejection Region')
    
    plt.title('F-Distribution and F-Test Result')
    plt.xlabel('F-statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(1)
data_set1 = np.random.normal(loc=100, scale=15, size=50)
data_set2 = np.random.normal(loc=105, scale=20, size=60)

f_test_and_visualize(data_set1, data_set2)

print("\nInterpretation of Visualization:")
print("The box plots show the spread and central tendency of each dataset. If the boxes have significantly different heights, it visually suggests different variances.")
print("The F-distribution plot shows the probability density for different F-statistic values. The calculated F-statistic is marked, along with the critical value(s) for the chosen alpha. If the calculated F-statistic falls into the shaded rejection region, then the null hypothesis of equal variances is rejected.")
```

### 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Scenario: A die roll experiment
# Null Hypothesis (H0): The die is fair (each face has an equal probability of landing).
# Alternative Hypothesis (H1): The die is not fair.

# Simulate observed frequencies from 120 rolls of a die
# Let's say we observed the following counts for faces 1 through 6:
observed_frequencies = np.array([15, 25, 18, 22, 20, 20]) # Sum = 120

# Calculate expected frequencies for a fair die
total_rolls = np.sum(observed_frequencies)
num_faces = len(observed_frequencies)
expected_frequencies = np.full(num_faces, total_rolls / num_faces)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies (Fair Die): {expected_frequencies}")

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed distribution is significantly different from the expected (fair) distribution.")
    print("Conclusion: The die is likely not fair.")
else:
    print("\nFail to reject the null hypothesis: The observed distribution does not significantly differ from the expected (fair) distribution.")
    print("Conclusion: There is no significant evidence to suggest the die is not fair.")

# Optional: Visualize the frequencies
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(num_faces)

width = 0.35
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, observed_frequencies, width, label='Observed', alpha=0.7)
plt.bar(x + width/2, expected_frequencies, width, label='Expected', alpha=0.7)

plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Observed vs. Expected Frequencies of Die Rolls')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
